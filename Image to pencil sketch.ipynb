```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
BGR_image = cv2.imread('hs.jpg')
```

```python
plt.imshow(BGR_image)
plt.title("BGR_Image",fontstyle='italic', fontsize=12,color="gray")

plt.axis(False)
plt.show()
```

```python
RGB_image = cv2.cvtColor(BGR_image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_image)
plt.title("RGB_Image",fontstyle='italic', fontsize=12,color="gray")
plt.axis(False)
plt.show()
```

```python
GreyScale_image = cv2.cvtColor(RGB_image, cv2.COLOR_RGB2GRAY)
plt.imshow(GreyScale_image)
plt.title("Grey_Scale_Image",fontstyle='italic', fontsize=12,color="gray")
plt.axis(False)
plt.show()
```

```python
img_smooth = cv2.cvtColor(RGB_image, cv2.COLOR_RGB2GRAY)

plt.imshow(img_smooth)
plt.axis(False)
plt.title('Smooth Image',fontstyle='italic', fontsize=12,color="gray")
plt.show()
```

```python
PencilSketch_image = cv2.cvtColor(RGB_image, cv2.COLOR_RGB2BGR)
plt.figure(figsize = (8,8))
plt.title("Pencil_Sketch_Image",fontstyle='italic', fontsize=12,color="gray")
plt.imshow(PencilSketch_image)
plt.axis(False)
plt.show()
```

```python
cv2.imwrite("sketch.jpg", PencilSketch_image)
```

```python
Comp = np.concatenate((RGB_image, PencilSketch_image), axis=1)
plt.figure(figsize = (8,8))
plt.imshow(Comp)
plt.axis(False)
plt.show()
```
